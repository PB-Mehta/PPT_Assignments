Given a Binary Tree (Bt), convert it to a Doubly Linked List(DLL). The left and right pointers in nodes are to be used as previous and next pointers respectively in converted DLL. The order of nodes in DLL must be the same as in Inorder for the given Binary Tree. The first node of Inorder traversal (leftmost node in BT) must be the head node of the DLL.

```python
def isSameTree(preorder, inorder, postorder):
    if len(preorder) != len(inorder) or len(inorder) != len(postorder):
        return False

    if not preorder:
        return True

    if preorder[0] != inorder[0] or inorder[0] != postorder[-1]:
        return False

    root_value = preorder[0]
    idx = inorder.index(root_value)

    left_preorder = preorder[1:idx+1]
    left_inorder = inorder[:idx]
    left_postorder = postorder[:idx]
    right_preorder = preorder[idx+1:]
    right_inorder = inorder[idx+1:]
    right_postorder = postorder[idx:-1]

    return isSameTree(left_preorder, left_inorder, left_postorder) and \
           isSameTree(right_preorder, right_inorder, right_postorder)
```

Given a binary tree, print all its root-to-leaf paths without using recursion. For example, consider the following Binary Tree.

```python
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None


def construct_tree(preorder, inorder):
    if not preorder or not inorder:
        return None

    root_val = preorder[0]
    root = TreeNode(root_val)

    # Find the root index in the inorder list
    root_index = inorder.index(root_val)

    # Construct the left and right subtrees recursively
    root.left = construct_tree(preorder[1:root_index + 1], inorder[:root_index])
    root.right = construct_tree(preorder[root_index + 1:], inorder[root_index + 1:])

    return root


def compare_postorder(root, postorder):
    if not root:
        return True

    # Compare left and right subtrees recursively
    left_check = compare_postorder(root.left, postorder)
    right_check = compare_postorder(root.right, postorder)

    # Check if the root value matches the next value in postorder traversal
    if root.val != postorder[0]:
        return False

    # Remove the matched root value from the postorder list
    postorder.pop(0)

    return left_check and right_check


def check_same_tree(preorder, inorder, postorder):
    # Construct the tree using the given preorder and inorder traversals
    root = construct_tree(preorder, inorder)

    # Check if the postorder traversal matches the constructed tree
    return compare_postorder(root, postorder)


# Example usage
preorder = [1, 2, 4, 5, 3, 6]
inorder = [4, 2, 5, 1, 6, 3]
postorder = [4, 5, 2, 6, 3, 1]

same_tree = check_same_tree(preorder, inorder, postorder)
print("The traversals are of the same tree:", same_tree)
```
